### Проверка правильности вычисленной эффективности обогащения

Напишем функцию для расчета Recovery по формуле

![jupyter](https://pictures.s3.yandex.net/resources/Recovery_1576238822.jpg)

где:

- C — доля золота в концентрате после флотации/очистки (rougher.output.concentrate_au)
- F — доля золота в сырье/концентрате до флотации/очистки (rougher.input.feed_au)
- T — доля золота в отвальных хвостах после флотации/очистки (rougher.output.tail_au)

```python
def calc_recovery(rougher_output, rougher_input, rougher_tail):
    recovery = (rougher_output * (rougher_input - rougher_tail)) /\
        (rougher_input * (rougher_output - rougher_tail)) * 100
    
    recovery[recovery < 0] = np.nan
    recovery[recovery > 100] = np.nan
    return recovery
```

Посчитаем `recovery` на обучающей выборке

```python
recovery = calc_recovery(
    df_train['rougher.output.concentrate_au'],
    df_train['rougher.input.feed_au'],
    df_train['rougher.output.tail_au']
)
print(recovery.describe())
```

```python
mae_recovery = mean_absolute_error(df_train['rougher.output.recovery'].dropna(), recovery.dropna())
print('MAE =', mae_recovery)
```

**Вывод:** 

MAE (средняя абсолютная ошибка) между рассчитаной по формуле эффективностью обогащения и признаком из датасета очень мала, а значит значение эффективности в датасете верное.

### Функция для вычисления итоговой sMAPE

![jupyter](https://pictures.s3.yandex.net/resources/smape_1576238825.jpg)

![jupyter](https://pictures.s3.yandex.net/resources/_smape_1576238814.jpg)

Напишем функцию для вычисления итоговой sMAPE

```python
def calculate_smape(target_actual, target_predictions):
    numerator = np.abs(target_actual - target_predictions)
    denominator = (np.abs(target_actual) + np.abs(target_predictions)) / 2 
    
    return 1 / len(target_actual) * np.sum(numerator / denominator) * 100 
    
    
def calculate_final_smape(target_actual, target_predictions):
    target_actual_rougher = target_actual['rougher.output.recovery']
    target_actual_final = target_actual['final.output.recovery']
    
    target_predictions_rougher = target_predictions[:, 0]
    target_predictions_final = target_predictions[:, 1]
    
    return 0.25 * calculate_smape(target_actual_rougher, target_predictions_rougher) \
        + 0.75 * calculate_smape(target_actual_final, target_predictions_final)
```

Создадим scorer для того, чтобы использовать его при кросс-валидации

```python
smape_scorer = make_scorer(calculate_final_smape, greater_is_better=False)
```

### Обучение моделей и оценка их качества кросс-валидацией

Выделим признаки и целевые показатели rougher.output.recovery и final.output.recovery

```python
features_train = df_train.drop(TARGETS, axis='columns')
target_train = df_train[TARGETS]
features_test = df_test.drop(TARGETS, axis='columns')
target_test = df_test[TARGETS]
```

Теперь обучим модели и посмотрим, какая даст лучшее значение sMAPE

#### Дерево решений

```python
param_grid = {'max_depth': range(2, 20)}

model = DecisionTreeRegressor(random_state=RANDOM_STATE)
search = GridSearchCV(model, param_grid, cv=5, scoring=smape_scorer)
search.fit(features_train, target_train)
print(
    'Лучшее значение sMAPE для дерева решений: {} при значениях гиперпараметров: {}'.format(
        -search.best_score_, search.best_params_)
)
```

#### Cлучайный лес

```python
param_grid = {'max_depth': range(2, 20)}

model = DecisionTreeRegressor(random_state=RANDOM_STATE)
search = GridSearchCV(model, param_grid, cv=5, scoring=smape_scorer)
search.fit(features_train, target_train)
print(
    'Лучшее значение sMAPE для дерева решений: {} при значениях гиперпараметров: {}'.format(
        -search.best_score_, search.best_params_)
)
```

<font color='blue'><b>Комментарий ревьюера: </b></font> ✔️\
<font color='green'>👍</font>

#### Линейная регрессия

```python
model = LinearRegression()
model.fit(features_train, target_train)
smape = cross_val_score(model, features_train, target_train, cv=5, scoring=smape_scorer).mean()
print('Лучшее значение sMAPE для линейной регрессии:', -smape)
```

Выяснили, что лучшая модель - **Cлучайный лес**. Проверим ее качество на тестовой выборке.

```python
model = RandomForestRegressor(max_depth=2, n_estimators=60, random_state=RANDOM_STATE) 
model.fit(features_train, target_train)
predictions_test = model.predict(features_test)

print('sMAPE для тестовых данных:', calculate_final_smape(target_test, predictions_test))
```

#### Проверим модель на адекватность

```python
dummy_model = DummyRegressor(strategy="mean")
dummy_model.fit(features_train, target_train)
test_predictions = dummy_model.predict(features_test)

print('Итоговое sMAPE для DummyRegressor:',calculate_final_smape(
    target_test, test_predictions).mean())
```

Минимальное значение sMape на тестовой выборке у случайного леса (max_depth=2, n_estimators=110) -- 9.44. Этот результат лучше среднего, модель вменяема.

**Выводы:**

- случайный лес дал лучшее значение sMAPE среди других моделей
- на тестовой выборке случайный лес предсказал значения целевых признаков rougher.output.recovery и final.output.recovery с sMAPE 9.44
