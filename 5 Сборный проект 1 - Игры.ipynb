## Предобработка данных

**Работа с пропусками:**

- пропуски в названиях игр заполнить не представляется возможным, оставим как есть, имеющаяся информация в этих строках понадобится при анализе данных;

### Заменим названия столбцов (приведем к нижнему регистру, проверим лишние пробелы).

```python
data.columns = data.columns.str.strip().str.lower()
data.columns
```

### Так же приведем к нижнему регистру и избавимся от лишних пробелов в значениях строковых столбцов, а так же сделаем проверку на явные дубликаты.

```python
data_columns = ['name', 'platform', 'genre', 'rating']

for col in data_columns:
    data[col] = data[col].str.strip().str.lower()
    
# проверим датафрейм на явные дубликаты
display(data.duplicated().value_counts())
data.drop_duplicates(subset=['name', 'platform', 'year_of_release'], keep='first')
```

Явных дубликатов нет.

### Преобразуем данные в нужные типы и обработаем пропуски при необходимости.

```python
# проверим строки в которых рейтинг пользователей заявлен как tbd
data.query('user_score == "tbd"')
```

Никакой связи между значением tbd и платформой, годом выпуска или продажами нет, в рейтинге ESRB это значение 
обозначает to be decided и связано с тем, что отзывов недостаточно для проставления оценки игре, эти оценки в датафрейме просто не были обновлены после появления реальной оценки от пользователей, можно заменить tbd на NaN.

```python
# заменим tbd на NaN, чтобы преобразовать тип данных в столбце
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.NaN
```

```python
# для замены типа заполним пропуски в year_of_release на маркерный год(выберем 2025)
# заменим тип данных в year_of_release на integer, изменять на .dt.year нет необходимости, 
# так как результат будет тот же при большем количестве кода
data['year_of_release'] = data['year_of_release'].fillna(2025).astype('int')

# изменим тип данных в user_score на float
data['user_score'] = data[data['user_score'].notnull()]['user_score'].astype('float')

data.info()
```

```python
data.drop(labels = [659, 14244], inplace=True)
```

```python
# Заполним пропуски в рейтинге маркерным значением
data['rating'] = data['rating'].fillna('n/a')
```

```python
# проверим датафрейм на явные дубликаты
data.duplicated().value_counts()
# явных дубликатов нет
```

```python
data_isna = data.isna().mean()*100
```

```python
def pass_value_barh(y, x):
    # Построим график
    # Размер графика
    fig, ax = plt.subplots(figsize=(16, 10))
 
    # Горизонтальная гистограмма
    ax.barh(y, x)
 
    # Добавим x, y гридлайны
    ax.grid(b=True, color='grey',
        linestyle='-.', linewidth=0.5,
        alpha=0.2)
 
    # Отсортируем по убыванию
    ax.invert_yaxis()
 
    # Добавим значения к столбцам гистограммы
    for i in ax.patches:
        plt.text(i.get_width()+0.2, i.get_y()+0.5,
                 str(round((i.get_width()), 2)),
                 fontsize=10, fontweight='bold',
                 color='grey')
    plt.xlabel('Количество')
    # Show Plot
    plt.show()
```

```python
pass_value_barh(data_isna.index, data_isna);
```

## Проверка гипотез

### Гипотеза 1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

**H0 (нулевая гипотеза)** - Средние пользовательские рейтинги платформ Xbox One и PC равны;                   

**H1 (альтернативная гипотеза)** - Средние пользовательские рейтинги платформ Xbox One и PC различаются между собой.

```python
xone_rating = data_actual.query('platform == "xone" & user_score > 0')['user_score']
pc_rating = data_actual.query('platform == "pc" & user_score > 0')['user_score']
alpha = 0.05
```

```python
display(xone_rating.var())
pc_rating.var()
```

```python
stat, p = st.ttest_ind(
    xone_rating, 
    pc_rating
)
print('t-статистика = {}\nНаименьшее значение уровня значимости p-value = {:.10f}'.format(stat, p))
print('\nКритический уровень статистической значимости alpha = 0.05')

if p < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
display(xone_rating.mean())
pc_rating.mean()
```

**Вывод:** 

**На имеющихся данных, на уровне значимости 5% (уровне доверия 95%) нет оснований отвергнуть нулевую гипотезу в пользу альтернативы. Полученное значение p-value говорит о том, что хотя средний рейтинг Xbox One и ПК неодинаков, с вероятностью в почти 14% такое или большее различие можно получить случайно. Это явно слишком большая вероятность, чтобы делать вывод о значимом различии между средними рейтингами.**

```python
# Приведены два датасета: сумма покупок, совершённых за месяц посетителями ...

sample_1 = [3071, 3636, 3454, 3151, 2185, 3259, 1727, 2263, 2015,
2582, 4815, 633, 3186, 887, 2028, 3589, 2564, 1422, 1785,
3180, 1770, 2716, 2546, 1848, 4644, 3134, 475, 2686,
1838, 3352]
sample_2 = [1211, 1228, 2157, 3699, 600, 1898, 1688, 1420, 5048, 3007,
509, 3777, 5583, 3949, 121, 1674, 4300, 1338, 3066,
3562, 1010, 2311, 462, 863, 2021, 528, 1849, 255,
1740, 2596]
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(
sample_1,
sample_2)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:

    print("Не получилось отвергнуть нулевую гипотезу")
```

### Гипотеза 2. Средние пользовательские рейтинги жанров Action (экшен-игры) и Sports (спортивные) разные.

**H0 (нулевая гипотеза)** - Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) равны средним пользовательским рейтингам жанра Sports (англ. «спортивные соревнования»).

**H1 (альтернативная гипотеза)** - Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) не равны средним пользовательским рейтингам жанра Sports (англ. «спортивные соревнования»).

```python
action_genre = data_actual.query('genre == "action" & user_score > 0')['user_score']
sports_genre = data_actual.query('genre == "sports" & user_score > 0')['user_score']
alpha = 0.05
```

```python
display(action_genre.var())
sports_genre.var()
```

```python
stat, p = st.ttest_ind(
    action_genre, 
    sports_genre,
)
print('t-статистика = {}\nНаименьшее значение уровня значимости p-value = {:.30f}'.format(stat, p))
print('\nКритический уровень статистической значимости alpha = 0.05')

if p < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
display(action_genre.mean())
sports_genre.mean()
```

**Вывод:** 

**На имеющихся данных, на уровне значимости 5% (уровне доверия 95%) есть основания отвергнуть нулевую гипотезу в пользу альтернативы. При проведении статистического теста по актуальному периоду, получилось, что средние рейтинги жанров action и sports не равны и нулевая гипотеза отвергается. Таким образом, при полученном наименьшем значении уровня значимости(p-value) 0.0000000000000000000000001052% у нас практически нет вероятности получить одинаковые средние рейнтинги по двум жанрам.**
